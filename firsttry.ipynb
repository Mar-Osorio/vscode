```python

import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril'],
    'Ventas': [1500, 1800, 1700, 2000]
}

df = pd.DataFrame(data)

plt.plot(df['Mes'], df['Ventas'], marker='o', color='pink')
plt.title('Ventas por Mes')
plt.xlabel('Mes')
plt.ylabel('Ventas ($)')
plt.tight_layout()
plt.show()

```
